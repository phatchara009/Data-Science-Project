# **Control Flow**

Transferable Knowledge

- if
- for loop

```r
## ifelse
## =IF() in google sheets
## ifelse() in R

score <- 88

ifelse(score >= 80, "passed", "failed")
```

```r
## if-else
score <- 62

if (score >= 80) {
    print("A")
} else if (score >= 70) {
    print("B")
} else if (score >= 60) {
    print("C")
} else if (score >= 50) {
    print("D")
} else {
    print("Failed")
}
```

```r
grading <- function(score) {
    if (score >= 80) {
        print("A")
    } else if (score >= 70) {
        print("B")
    } else if (score >= 60) {
        print("C")
    } else if (score >= 50) {
        print("D")
    } else {
        print("Failed")
    }
}
```

```r
grading(100)
```

```r
# infinite loop
# while (1 > 0) {
#     print("hi!")
# }
count <- 0
while (1 > 0) {
    print("hi!")
    count = count + 1
    if (count == 5) {
        print("Done!")
        break
    }
}
```
