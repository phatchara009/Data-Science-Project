# Rock-Paper-Scissors game

```r
game <- function() {

    hands <- 1:3
    score_player <- 0
    score_comp <- 0

    convertNumberToAction <- function(number) {
        if (number == 1) {
            x = "Hammer"
        } else if (number == 2) {
            x = "Paper"
        } else if (number == 3) {
            x = "Scissors"
        }
        return (x)
    }

    print("Hello Player, Welcome to the Rock-Paper-Scissors game!!!")

    cat("\n")
    print("Please enter the number of Rock-Paper-Scissors")
    print("1 = Hammer")
    print("2 = Paper")
    print("3 = Scissors")
    print("99 = Quit")

    cat("\n")
    print("If you win, giving you 1 point")

    cat("\n")
    print("You can play as many turns as you want.")
    print("If you want to stop playing games")
    print("Please enter the command: 99")

    cat("\n")
    print("Ok, Let's start Rock-Paper-Scissors")

    cat("\n")

    player_hand <- 0

    while (player_hand != 99) {

        flush.console()
        player_hand <- as.numeric(readline("\nChoose your hand (number): "))

        ## Quit and Error
        if (player_hand == 99) {
            break
        } else if (!(player_hand %in% c(1, 2, 3, 99))) {
            print("Please only enter 1, 2, 3, or 99")
            next
        }

        comp_hand <- sample(hands,1)

        condition_win <- (player_hand == 1 & comp_hand == 3) | (player_hand == 2 & comp_hand == 1) | (player_hand == 3 & comp_hand == 2)
        condition_lost <- (player_hand == 3 & comp_hand == 1) | (player_hand == 1 & comp_hand == 2) | (player_hand == 2 & comp_hand == 3)
        condition_draw <- (player_hand == 1 & comp_hand == 1) | (player_hand == 2 & comp_hand == 2) | (player_hand == 3 & comp_hand == 3)

        print(paste("Player hand is", convertNumberToAction(player_hand)))
        print(paste("Computer hand is", convertNumberToAction(comp_hand)))

        ## Player Win and add player score
        if (condition_win) {
            score_player <- score_player + 1
            print("You win this turn, giving you 1 point.")

        ## Player Lost and add computer score
        } else if (condition_lost) {
            score_comp <- score_comp + 1
            print("You lost this turn, giving computer 1 point.")

        ## Player draw wih Computer
        } else if (condition_draw) {
            print("We draw this turn, no one scores")
        }

    }

    cat("\n")
    print(paste("Your score is", score_player))
    print(paste("Computer score is", score_comp))

    cat("\n")
    if (score_player > score_comp) {
        print("You Win!! :)")
    } else if (score_player < score_comp) {
        print("You Lost :(")
    } else {
        print("You draw with Computer :|")
    }

}
```

```r
game()
```
