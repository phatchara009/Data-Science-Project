# **Rock-Paper-Scissors Game**

```python
def game():

    import random
    # create function to convert number to action of this game
    def convertNumberToAction(number):
        if number == 1:
            x = "Hammer"
        elif number == 2:
            x = "Paper"
        elif number == 3:
            x = "Scissors"
        return x

    print("Hello Player, Welcome to the Rock-Paper-Scissors game!!!\n")

    print("Please enter the number of Rock-Paper-Scissors")
    print("1 = Hammer")
    print("2 = Paper")
    print("3 = Scissors")
    print("99 = Quit")

    print("\nIf you win, giving you 1 point")

    print("\nYou can play as many turns as you want.")
    print("If you want to stop playing games")
    print("Please enter the command: 99")

    print("\nOk, Let's start Rock-Paper-Scissors Game!!")

    # set initial value.
    score_player = 0
    score_comp = 0

    # start the game
    while True:

        player_hand = int(input("\nChoose your hand (1, 2, 3, or 99): "))

        ## Quit and Error
        if player_hand == 99:
            print("\nGame has stop!!")
            break
        elif (player_hand > 3 and player_hand < 1):
            print("\nPlease only enter 1, 2, 3, or 99")
            continue
        elif (player_hand >= 1 and player_hand <= 3):

            comp_hand = random.randint(1,3)
            player = convertNumberToAction(player_hand)
            comp = convertNumberToAction(comp_hand)

            print(f"Player hand is {player}.")
            print(f"Computer hand is {comp}.")

            ## Player Win and add player score
            if ((player_hand == 1 and comp_hand == 3) or (player_hand == 2 and comp_hand == 1) or (player_hand == 3 and comp_hand == 2)):
                score_player += 1
                print("You win this turn, giving you 1 point.")

            ## Player Lost and add computer score
            elif ((player_hand == 3 and comp_hand == 1) or (player_hand == 1 and comp_hand == 2) or (player_hand == 2 and comp_hand == 3)):
                score_comp += 1
                print("You lost this turn, giving computer 1 point.")

            ## Player draw wih Computer
            elif ((player_hand == 1 and comp_hand == 1) or (player_hand == 2 and comp_hand == 2) or (player_hand == 3 and comp_hand == 3)):
                print("We draw this turn, no one scores")

    if (score_player == 0 and score_comp == 0):
        print("\nWhy don't you start playing the game? I'm waiting to play with you <3.")
    elif (score_player > score_comp):
        print(f"\nYour score is {score_player} point.")
        print(f"Computer score is {score_comp} point.\n")
        print("You Win!! :)")
    elif (score_player < score_comp):
        print(f"\nYour score is {score_player} point.")
        print(f"Computer score is {score_comp} point.\n")
        print("You Lost :(")
    else:
        print(f"\nYour score is {score_player} point.")
        print(f"Computer score is {score_comp} point.\n")
        print("You draw with Computer :|")
```

```python
game()
```
