```python
import pandas as pd
pd.set_option("display.precision", 2)
```

```python
df = pd.read_excel("data/finance_data.xlsx")
df.head()
```

```python
df_transposed = df.set_index("Indicator Name").transpose().reset_index(names=["Date"])
df_transposed.rename_axis(None, axis=1, inplace=True)
df_transposed.tail()
```

```python
# Drop Monthly Financial Ratio
df_transposed["Date"] = pd.to_datetime(df_transposed['Date'])
```

```python
df_transposed.info()
```

```python
df_transposed.to_csv("data/finance_data_clean")
```
