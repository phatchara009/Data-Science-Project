```python
# Question 1-2:

# Write a program that prompts the user to enter a number and calculates 
# the factorial of that number using a while loop. Assume the input is a positive integer.

# Factorial is a mathematical operation that calculates the product of an integer and
# all the positive integers below it. Example: 5! = 5 x 4 x 3 x 2 x 1 = 120.

num = int(input("Enter a positive integer: "))
factorial = 1
while num > 0:
    factorial *= num
    num -= 1
print("Factorial:", factorial)
```

```python
# Question 1-3:

# Write a program that asks the user to enter a positive integer and 
# prints whether the number is prime or not using a while loop.

# Prime number is a natural number greater than 1 that is divisible only by 1 and itself.
# Example: 7 is a prime number because it is only divisible by 1 and 7, 
# whereas 6 is not a prime number because it is divisible by 1, 2, 3, and 6 in addition to itself.

num = int(input("Enter a positive integer: "))
is_prime = True
divisor = 2
while divisor <= num // 2:
    if num % divisor == 0:
        is_prime = False
        break
    divisor += 1
if is_prime:
    print(num, "is prime.")
else:
    print(num, "is not prime.")
```

```python
# Example 2-1: Fibonacci Sequence

# Generate and print Fibonacci sequence up to a given number of terms

# The Fibonacci sequence is a series of numbers where each number is the sum of the two preceding ones.
# Example: The Fibonacci sequence starts with 0 and 1, and 
# each subsequent number is the sum of the previous two numbers. The sequence begins as follows: 0, 1, 1, 2, 3, 5, 8, 13, 21, and so on.

terms = int(input("Enter the number of terms: "))

# First two terms
a, b = 0, 1
count = 0

while count < terms:
    print(a)
    temp = a + b
    a = b
    b = temp
    count += 1
```

```python
# Example 2-2: Guessing Game

# Guessing game where the user has to guess a randomly generated number

import random

target_number = random.randint(1, 100)
guess = 0
attempts = 0

while guess != target_number:
    guess = int(input("Guess the number (between 1 and 100): "))
    attempts += 1

    if guess < target_number:
        print("Too low!")
    elif guess > target_number:
        print("Too high!")

print("Congratulations! You guessed the number in", attempts, "attempts.")
```

```python
# Question 2-1:

# Write a program that asks the user to enter a positive integer 
# and calculates the sum of all the numbers from 1 to that integer using a while loop.

num = int(input("Enter a positive integer: "))
sum = 0
count = 1

while count <= num:
    print("count =", count)
    print("sum =", sum)
    sum += count
    count += 1
    print("----") # sum(N+1) = sum(N) + count(N)
    
print("Sum:", sum)
```

```python
# Question 2-2:

# Write a program that prompts the user to enter a positive integer 
# and determines whether it is a palindrome number using a while loop. 
# A palindrome number reads the same forwards and backward.

# A palindrome number is a number that remains the same when its digits are reversed.
# Example: 121 is a palindrome number because it reads the same backward as forward. However, 123 is not a palindrome number because it reads differently when its digits are reversed (321).

# https://www.mathmyself.com/G/13/%E0%B8%88%E0%B9%8D%E0%B8%B2%E0%B8%99%E0%B8%A7%E0%B8%99%E0%B8%9E%E0%B8%B2%E0%B8%A5%E0%B8%B4%E0%B8%99%E0%B9%82%E0%B8%94%E0%B8%A3%E0%B8%A1(Palindrome-Numbers).htm


num = int(input("Enter a positive integer: "))
original_num = num
reverse_num = 0

while num > 0:
    remainder = num % 10
    reverse_num = (reverse_num * 10) + remainder
    num = num // 10

if original_num == reverse_num:
    print(original_num, "is a palindrome number.")
else:
    print(original_num, "is not a palindrome number.")
```

```python
# Example 3-1: Prime Number Generator

# A prime number generator is a tool or algorithm that generates a sequence of prime numbers.
# Example: The Sieve of Eratosthenes is a prime number generator algorithm that 
# efficiently generates all prime numbers up to a given limit by iteratively sieving out the multiples of each prime number.

# Generate and print prime numbers up to a given limit

limit = int(input("Enter the upper limit: "))

num = 2
while num <= limit:
    is_prime = True
    divisor = 2

    while divisor <= num // 2:
        if num % divisor == 0:
            is_prime = False
            break
        divisor += 1

    if is_prime:
        print(num)
    num += 1
```

```python
# Example 3-2: Factorial Calculation

# Factorial calculation is the process of finding the product of an integer and all the positive integers below it.
# Example: The factorial of 5, denoted as 5!, is calculated by multiplying 5 by all the positive integers below it: 5! = 5 x 4 x 3 x 2 x 1 = 120.

# Calculate the factorial of a given number

num = int(input("Enter a positive integer: "))
factorial = 1

while num > 1:
    factorial *= num
    num -= 1

print("Factorial:", factorial)
```

```python
# Example 3-3: Binary Conversion

# Binary conversion is the process of converting a number from decimal (base 10) to binary (base 2) representation.
# Example: The decimal number 10 can be converted to binary by repeatedly dividing it by 2 and 
# noting the remainders: 10 / 2 = 5 remainder 0, 5 / 2 = 2 remainder 1, 2 / 2 = 1 remainder 0. The binary representation of 10 is 1010.

# Convert a decimal number to its binary representation
# https://www.cuemath.com/numbers/decimal-to-binary/

decimal = int(input("Enter a decimal number: "))
binary = ""

while decimal > 0:
    binary = str(decimal % 2) + binary
    decimal //= 2

print("Binary:", binary)
```

## Answer for level 3

```python
# Question 3-1:

# Write a program that asks the user to enter a positive integer and 
# checks if it is a perfect number using a while loop. 
# A perfect number is a positive integer that is equal to the sum of its proper divisors.

# A perfect number is a positive integer that is equal to the sum of its proper divisors (excluding the number itself).
# Example: The number 6 is a perfect number because its proper divisors (excluding 6) are 1, 2, and 3, and their sum is equal to 6: 1 + 2 + 3 = 6.

# จำนวนสมบูรณ์ (perfect number) คือ จำนวนเต็มที่มีผลบวกของตัวหารแท้ เท่ากับตัวมันเอง
# หก (6) เป็นจำนวนสมบูรณ์ตัวแรก เพราะว่า 6 มีตัวหารแท้คือ 1, 2, 3

num = int(input("Enter a positive integer: "))
divisor = 1
sum_of_divisors = 0

while divisor < num:
    if num % divisor == 0:
        sum_of_divisors += divisor
    divisor += 1

if sum_of_divisors == num:
    print(num, "is a perfect number.")
else:
    print(num, "is not a perfect number.")
```

```python
# Question 3-2:

# Write a program that prompts the user to enter a positive integer and 
# determines whether it is an Armstrong number using a while loop.

# An Armstrong number is a number that is equal to the sum of its own digits raised to the power of the number of digits.
# Example: The number 153 is an Armstrong number because it has 3 digits, and 
# the sum of its digits raised to the power of 3 is equal to the number itself: 1^3 + 5^3 + 3^3 = 153.

# An Armstrong number is a number that is equal to the sum of its own digits, 
# each raised to the power of the number of digits.

num = int(input("Enter a positive integer: "))
original_num = num
num_of_digits = len(str(num))
sum_of_digits = 0

while num > 0:
    digit = num % 10
    sum_of_digits += digit ** num_of_digits
    num //= 10

if original_num == sum_of_digits:
    print(original_num, "is an Armstrong number.")
else:
    print(original_num, "is not an Armstrong number.")
```

```python
# Question 3-3:

# Write a program that prompts the user to enter a positive integer and 
# prints its reverse using a while loop.

# input = 567 => output = 765

num = int(input("Enter a positive integer: "))
reverse = 0

while num > 0:
    digit = num % 10
    reverse = (reverse * 10) + digit
    num //= 10

print("Reverse:", reverse)
```

```python
# Question 3-2:

# Write a program that prompts the user to enter a string and 
# counts the occurrences of each character in the string using a for loop. 
# Print the characters along with their counts.

string = input("Enter a string: ")
character_count = {}

for char in string:
    if char in character_count:
        character_count[char] += 1
    else:
        character_count[char] = 1

print("Character counts:")
for char, count in character_count.items():
    print(f"{char}: {count}")
```

```python
# Question 3-3:

# Write a program that prompts the user to enter a number and checks if it is a prime number using a for loop. 
# A prime number is a number greater than 1 that is divisible only by 1 and itself.

number = int(input("Enter a number: "))
is_prime = True

if number > 1:
    for i in range(2, number):
        if number % i == 0:
            is_prime = False
            break
else:
    is_prime = False

if is_prime:
    print(number, "is a prime number.")
else:
    print(number, "is not a prime number.")
```

```python
# Example 4-1: Generate Fibonacci Sequence

# Generate the Fibonacci sequence up to a given number of terms

terms = int(input("Enter the number of terms: "))

fibonacci = [0, 1]

for i in range(2, terms):
    next_number = fibonacci[i-1] + fibonacci[i-2]
    fibonacci.append(next_number)

print("Fibonacci sequence:")
for number in fibonacci:
    print(number, end=" ")
```

```python
# Print a pyramid pattern using a for loop

size = 5

for i in range(1, size + 1):
    print(" " * (size - i) + "*" * (2 * i - 1))
```

```python
# Write a program that generates a list of prime numbers up to a given limit using a for loop. 
# A prime number is a number greater than 1 that is divisible only by 1 and itself.

limit = int(input("Enter the limit: "))
prime_numbers = []

for number in range(2, limit + 1):
    is_prime = True
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            is_prime = False
            break
    if is_prime:
        prime_numbers.append(number)

print("Prime numbers up to", limit, ":")
print(prime_numbers)
```
