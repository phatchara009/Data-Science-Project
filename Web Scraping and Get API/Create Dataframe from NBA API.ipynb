```python
import pandas as pd
import numpy as np
from requests import get
from time import sleep
```

## Get All Players

```python
url = "https://free-nba.p.rapidapi.com/players/"

players = pd.DataFrame(columns=["first_name", "last_name", "height_feet", "position", "team_name", "team_city"])

headers = {
	"X-RapidAPI-Key": "caaaf2a594msh67229ea59deac1fp10efc5jsn62f662319b85",
	"X-RapidAPI-Host": "free-nba.p.rapidapi.com"
}

i = 0
while True:
    i = i + 1
    index = str(i)
    new_url = url + index
    try:
        resp = get(new_url, headers=headers).json()
        players.loc[index, "first_name"] = resp["first_name"]
        players.loc[index, "last_name"] = resp["last_name"]
        players.loc[index, "height_feet"] = resp["height_feet"]
        players.loc[index, "position"] = resp["position"]
        players.loc[index, "team_code"] = resp["team"]["abbreviation"]
        sleep(0.5) # stop loop 0.5 second
    except:
        break

players.reset_index(drop=True, inplace=True)
```

## Get All Teams

```python
url = "https://free-nba.p.rapidapi.com/teams/"

teams = pd.DataFrame(columns=["abbreviation", "city", "conference", "division", "full_name", "name"])

headers = {
    "X-RapidAPI-Key": "caaaf2a594msh67229ea59deac1fp10efc5jsn62f662319b85",
    "X-RapidAPI-Host": "free-nba.p.rapidapi.com"
}

i = 0
x = 0

while True:
    i = i + 1
    index = str(i)
    new_url = url + index
    try:
        resp = get(new_url, headers=headers).json()
        teams.loc[index, "abbreviation"] = resp["abbreviation"]
        teams.loc[index, "city"] = resp["city"]
        teams.loc[index, "conference"] = resp["conference"]
        teams.loc[index, "division"] = resp["division"]
        teams.loc[index, "full_name"] = resp["full_name"]
        teams.loc[index, "name"] = resp["name"]
        sleep(1) # stop loop 1 second
    except:
        x = x + 1
        if x < 10:
            continue
        else:
            break
teams.reset_index(drop=True, inplace=True)
```

## Get All Games

```python
url = "https://free-nba.p.rapidapi.com/games/"

games = pd.DataFrame(columns=["date", "home_team", "home_team_score", "period", "postseason", "season", "status", "time", "visitor_team", "visitor_team_score"])

headers = {
    "X-RapidAPI-Key": "caaaf2a594msh67229ea59deac1fp10efc5jsn62f662319b85",
    "X-RapidAPI-Host": "free-nba.p.rapidapi.com"
}

i = 0
x = 0

while True:
    i = i + 1
    index = str(i)
    new_url = url + index
    try:
        resp = get(new_url, headers=headers).json()
        games.loc[index, "date"] = resp["date"]
        games.loc[index, "home_team"] = resp["home_team"]["abbreviation"]
        games.loc[index, "home_team_score"] = resp["home_team_score"]
        games.loc[index, "period"] = resp["period"]
        games.loc[index, "postseason"] = resp["postseason"]
        games.loc[index, "season"] = resp["season"]
        games.loc[index, "status"] = resp["status"]
        games.loc[index, "time"] = resp["time"].strip()
        games.loc[index, "visitor_team"] = resp["visitor_team"]["abbreviation"]
        games.loc[index, "visitor_team_score"] = resp["visitor_team_score"]
        sleep(0.25) # stop loop 1 second
    except:
        x = x + 1
        if x < 10:
            continue
        else:
            break
games.reset_index(drop=True, inplace=True)
```
