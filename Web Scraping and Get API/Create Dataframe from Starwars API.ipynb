# Project - Create Dataframe from Starwars API

```python
import pandas as pd
import numpy as np
from requests import get
from time import sleep
```

```python
url = "https://swapi.dev/api/people/"

starwars_char = pd.DataFrame(columns=["name", "species", "height", "gender", "homeworld", "lastest_movie"])

x = 0
for i in range(82):
    index = str(i + 1)
    new_url = url + index
    resp = get(new_url).json()
    try:
        starwars_char.loc[index, "name"] = resp["name"]
        try:
            starwars_char.loc[index, "species"] = get(resp["species"][0]).json()["name"]
        except:
            starwars_char.loc[index, "species"] = np.nan
        starwars_char.loc[index, "height"] = int(resp["height"])
        starwars_char.loc[index, "gender"] = "Genderless" if resp["gender"] == "n/a" else resp["gender"].capitalize()
        starwars_char.loc[index, "homeworld"] = get(resp["homeworld"]).json()["name"]
        starwars_char.loc[index, "lastest_movie"] = get(resp["films"][-1]).json()["title"]
        sleep(1) # stop loop 1 second
    except:
        x = x + 1
        if x < 10:
            continue
        else:
            break

starwars_char.reset_index(drop=True, inplace=True)
```
