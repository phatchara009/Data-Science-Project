```python
import pandas as pd
```

```python
set_stock_details = pd.read_csv('set_stock_details.csv')
set_stock_details.tail()
```

```python
mai_stock_details = pd.read_csv('mai_stock_details.csv')
mai_stock_details.tail()
```

```python
all_stock_details = pd.concat([set_stock_details, mai_stock_details], ignore_index=True)
all_stock_details.tail()
```

```python
all_stock_details.to_csv('all_stock_details.csv', index=False, encoding='utf-8-sig')
```
