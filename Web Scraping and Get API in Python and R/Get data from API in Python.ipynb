```python
from requests import get

url = "https://swapi.dev/api/people/"

for i in range(10):
    index = i + 1
    new_url = url + str(index)
    resp = get(new_url)
    print(resp.json()["name"])
```

```python
from requests import get
from time import sleep

url = "https://swapi.dev/api/people/"

for i in range(10):
    index = i + 1
    new_url = url + str(index)
    resp = get(new_url).json()
    name = resp["name"]
    height = resp["height"]
    print(name, height)
    sleep(2) # sleep 2 second
```
