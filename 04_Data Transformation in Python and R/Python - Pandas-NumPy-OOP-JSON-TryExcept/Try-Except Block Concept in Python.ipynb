# Try Except Block

```python
# ## Error divide by zero
# 1/0
```

```python
try:
    1/0
except:
    print("cannot divide by zero")
```

```python
try:
    1/0
except ZeroDivisionError:
    print("cannot divide by zero")
except NameError:
    print("variable not defined")
```

```python
try:
    number/0
except ZeroDivisionError:
    print("cannot divide by zero")
except NameError:
    print("variable not defined")
```

```python
try:
    1/5
except ZeroDivisionError:
    print("cannot divide by zero")
except NameError:
    print("variable not defined")
else:
    print("Done")
```

```python
try:
    1/5
except ZeroDivisionError:
    print("cannot divide by zero")
except NameError:
    print("variable not defined")
else:
    print("Done")
finally:
    print("Complete!")
```
