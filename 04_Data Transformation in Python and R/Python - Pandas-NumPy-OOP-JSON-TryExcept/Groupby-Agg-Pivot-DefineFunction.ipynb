# 1) Group by

```python
data = {
    'Category': ['Electronics', 'Electronics', 'Clothing', 'Clothing', 'Books', 'Books', 'Electronics', 'Electronics', 'Clothing', 'Clothing', 'Books', 'Books'],
    'Item': ['Laptop', 'Smartphone', 'T-shirt', 'Jeans', 'Novel', 'Textbook', 'Headphones', 'Tablet', 'Dress', 'Shoes', 'Magazine', 'Dictionary'],
    'Price': [1200, 800, 20, 60, 10, 50, 1000, 900, 40, 80, 5, 20],
    'Quantity': [5, 10, 50, 20, 100, 30, 8, 12, 30, 15, 50, 40],
    'City': ['Bangkok', 'Bangkok', 'Bangkok', 'Bangkok', 'Bangkok', 'Bangkok', 'Chiangmai', 'Chiangmai', 'Chiangmai', 'Chiangmai', 'Chiangmai', 'Chiangmai']
}

df = pd.DataFrame(data)
df
```

## Question 1: Given a DataFrame with sales data for different categories and cities, find the total revenue for each category.

```python
category_revenue = df.groupby('Category')['Price'].sum()
category_revenue
```

## Question 2: Given a DataFrame with sales data for different categories and cities, find the total revenue and quantity sold for each category and city.

```python
category_city_data = df.groupby(['Category', 'City']).agg({'Price': 'sum', 'Quantity': 'sum'})
category_city_data
```

# Question 3: Given a DataFrame with sales data for different categories and cities, find the average price of the most expensive item sold in each category and city combination.

```python
# First, sort the DataFrame by price within each category and city combination
df_sorted = df.sort_values(['Category', 'City', 'Price'], ascending=[True, True, False])

# Then, group the sorted DataFrame by category and city and take the mean of the first (i.e., most expensive) item in each group
category_city_avgprice = df_sorted.groupby(['Category', 'City']).apply(lambda x: x.head(1)['Price'].mean())
category_city_avgprice
```

```python
category_city_avgprice_df = category_city_avgprice.to_frame().reset_index()
category_city_avgprice_df.columns = ['Category', 'City', 'AvgPrice']
category_city_avgprice_df
```

# Question 4: Given a DataFrame with student grades for different subjects, find the average grade for each subject.

```python
data = {
    'Subject': ['Math', 'Math', 'English', 'English', 'Science', 'Science'],
    'Grade': [80, 90, 70, 85, 75, 80],
    'Student': ['John', 'Alice', 'John', 'Alice', 'John', 'Alice']
}

df = pd.DataFrame(data)
df
```

# Question 5: Given a DataFrame with sales data for different categories and cities, find the total revenue for each category and city combination.

```python
data = {
    'Category': ['Electronics', 'Electronics', 'Clothing', 'Clothing', 'Books', 'Books', 'Electronics', 'Electronics', 'Clothing', 'Clothing', 'Books', 'Books'],
    'Item': ['Laptop', 'Smartphone', 'T-shirt', 'Jeans', 'Novel', 'Textbook', 'Headphones', 'Tablet', 'Dress', 'Shoes', 'Magazine', 'Dictionary'],
    'Price': [1200, 800, 20, 60, 10, 50, 1000, 900, 40, 80, 5, 20],
    'Quantity': [5, 10, 50, 20, 100, 30, 8, 12, 30, 15, 50, 40],
    'City': ['Bangkok', 'Bangkok', 'Bangkok', 'Bangkok', 'Bangkok', 'Bangkok', 'Chiangmai', 'Chiangmai', 'Chiangmai', 'Chiangmai', 'Chiangmai', 'Chiangmai']
}

df = pd.DataFrame(data)
df.head(3)
```

```python
df['Revenue'] = df['Price'] * df['Quantity']
df.head(3)
```

```python
q122 = df.groupby(['Category', 'City'])["Revenue"].sum()
print(q122)

q122 = df.groupby(['Category', 'City']).agg({'Revenue': 'sum'})
print(q122)
```

## Question 1: Create a pivot table from the given DataFrame to show the total sales by city and product.

```python


result = pd.pivot_table(df, values='Sales', index=['City'], columns=['Product'], aggfunc='sum')
result
```

## Question 2: Create a pivot table to show the average sales by product and date, and add a column to show the percentage change in sales from the previous date for each product.

```python
result = pd.pivot_table(df, values='Sales', index=['Product', 'Date'], aggfunc='mean')
result
```

```python
# Add Sales Difference from previos date
result['SalesDiff'] = result.groupby(['Product'])['Sales'].pct_change()
result
```

## Question 3: Create a pivot table to show the total sales by city and product, and add a column to show the percentage of the total sales contributed by each product for each city.

```python
result = pd.pivot_table(df, values='Sales', index=['City'], columns=['Product'], aggfunc='sum')
result['TotalSales'] = result.sum(axis=1) # sum ออกไปแนว row
result
```

```python
result = result.apply(lambda x: x / x['TotalSales'], axis=1)
result = result.drop(columns=['TotalSales'])
result
```

## Question 2: Convert the following code that checks whether a number is prime or not into a Python function called is_prime:

```python
num = 17

if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
        else:
            print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

## Question 3: Convert the following code that computes the factorial of a number into a Python function called calculate_factorial

```python
num = 5

factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i
print("The factorial of", num, "is", factorial)
```

```python
# convert to function for question 3
def calculate_factorial(num):
    factorial = 1
    for i in range(1, num + 1):
        factorial = factorial * i
    return factorial

x = 7
print("The factorial of", num, "is", calculate_factorial(x))
```
